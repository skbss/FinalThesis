```python
xgb_e=create_model('xgboost')
```

```python
gbc_e=create_model('gbc')
```

```python
blend_soft = blend_models(estimator_list = [xgb_e, mlp,ada_e,RF_e,gbc_e],
                          optimize = 'Accuracy',method = 'soft')
```

```python
blend_soft = blend_models(estimator_list = [xgb_e, mlp,ada_e,RF_e,gbc_e],
                          optimize = 'Recall',method = 'soft')
```

```python
plot_model(blend_soft, plot='confusion_matrix')
```

```python
blend_hard = blend_models(estimator_list = [xgb_e, mlp,ada_e,RF_e,gbc_e],
                          optimize = 'Accuracy',method = 'hard')
```

```python
plot_model(blend_hard, plot='confusion_matrix')
```

```python
stack_model = stack_models(estimator_list = [xgb_e,ada_e,mlp,gbc_e], meta_model = RF_e,optimize = 'Recall')
```

```python
plot_model(stack_model, plot='confusion_matrix')
```
